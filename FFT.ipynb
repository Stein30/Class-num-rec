```python
#Twitter 

#sin(i/13)sin(i/2)
%matplotlib inline
from scipy import fft
import matplotlib.pyplot as plt
import numpy as np


def func(x):
    return np.sin(x/13)+np.sin(x/2)


x = np.linspace(0, 256, 256)  #02100numpy
#plt.plot(x, func(x), color = 'b')
plt.plot(x, np.sin(x/13), color = 'r', linewidth=0.8)
plt.plot(x, np.sin(x/2), color = 'b', linewidth=0.8)

plt.grid()
plt.show()

plt.plot(x, func(x), color = 'b')
plt.show()

yy = func(x)
out = fft(yy)

def spectrum_power(x):
    re, im = x.real, x.imag
    return np.sqrt(re**2+im**2)

plt.plot(x,spectrum_power(out))
plt.xlim(0,128)
plt.show()
```

```python
#2sin(i2)+sin(i/2)
%matplotlib inline
from scipy import fft
import matplotlib.pyplot as plt
import numpy as np


def func(x):
    return 2*np.sin(x*2)+np.sin(x/2)


x = np.linspace(0, 256, 256)  #02100numpy
#plt.plot(x, func(x), color = 'b')
plt.plot(x, 2*np.sin(x*2), color = 'r', linewidth=0.8)
plt.plot(x, np.sin(x/2), color = 'b', linewidth=0.8)

plt.grid()
plt.show()

plt.plot(x, func(x), color = 'b')
plt.show()

yy = func(x)
out = fft(yy)

def spectrum_power(x):
    re, im = x.real, x.imag
    return np.sqrt(re**2+im**2)

plt.plot(x,spectrum_power(out))
plt.xlim(0,128)
plt.show()
```
